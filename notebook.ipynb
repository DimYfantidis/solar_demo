#### Representation

* Each celestial body is represented as a sphere, $(x - x_i)^2 + (y - y_i)^2 + (z - z_i)^2 = R_i^2$, where $(x_i, y_i, z_i)$ and $R_i$ the position and the radius of the $i$-th planet, respectively.
* Let's siimplify the position of the planet using $p_i$, where $p_i = (x_i, y_i, z_i)$. This position is not static but changes as the time passes. Thus, position should be expressed as a function of time:
$$
    p_i(t) = (x_i(t),\:y_i(t),\:z_i(t))
$$
* Let $d_i$ be the distance between the sun and the $i-$th planet (in AU).
* Let $p_{sun}(t) = (0,\:0,\:0)$, then $p_i(t) = d_i \cdot (sin(t), cos(t), 0) \Rightarrow p_i(t) = (d_i sin(t), d_i cos(t), 0)$
